```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define the features to use
X_features = ['Gender', 'Status', 'Education', 'EmploymentStatus', 'MaritalStatus', 'HomeOwnershipType', 'UseOfLoan', 'Amount', 'Interest', 'LoanDuration', 'IncomeTotal', 'LiabilitiesTotal', 'DebtToIncome', 'FreeCash']
Y_Regression_features = ['EMI', 'ELA', 'ROI']

# Split the data into training and testing sets
X_train = df[X_features].loc[y_train.index]
X_test = df[X_features].loc[y_test.index]
y_train = df[Y_Regression_features].loc[y_train.index]
y_test = df[Y_Regression_features].loc[y_test.index]

# create an instance of the model
model = LinearRegression()

# create a pipeline
pipeline = Pipeline(steps=[('model', model)])

# fit the pipeline to the training data
pipeline.fit(X_train, y_train)

# make predictions on the test data
y_pred = pipeline.predict(X_test)

# evaluate the pipeline using metrics such as mean squared error (MSE) and R-squared
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean squared error:", mse)
print("R-squared:", r2)

# save the pipeline using pickle
import pickle

with open('linear_regression_pipeline.pkl', 'wb') as f:
    pickle.dump(pipeline, f)
```

#  Random Forest Classification

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Define the columns to be preprocessed using one-hot encoding
categorical_cols = ['Gender', 'Status', 'Education', 'EmploymentStatus', 'MaritalStatus', 'HomeOwnershipType', 'UseOfLoan']

# Define the columns to be preprocessed using standard scaling
numeric_cols = ['Amount', 'Interest', 'LoanDuration', 'IncomeTotal', 'LiabilitiesTotal', 'DebtToIncome', 'FreeCash']

# Define the preprocessor to apply the transformations
preprocessor_clf = ColumnTransformer(transformers=[('num', StandardScaler(), numeric_cols),
                                                   ('cat', OneHotEncoder(), categorical_cols)])

# Split the data into training and testing sets
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(df[X_features], df['Target'], test_size=0.2, random_state=42)

# Define the classifier model
model_clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Define the pipeline
pipeline_clf = Pipeline(steps=[('preprocessor', preprocessor_clf),
                               ('model', model_clf)])

# Fit the pipeline on the training data
pipeline_clf.fit(X_train_clf, y_train_clf)

# Make predictions on the test data
y_pred_clf = pipeline_clf.predict(X_test_clf)

# Calculate the accuracy score
accuracy = accuracy_score(y_test_clf, y_pred_clf)
print("Accuracy:", accuracy)
```
